```python
data2.sort_values(by = ['Salary'],ascending = False)
```

```python
def categorize_performance(rating):
  if rating >= 4.5:
    return 'Excellent'
  elif rating >= 4.0:
    return 'Good'
  else:
    return 'Average'

data2['Performance Category'] = data2['Rating'].apply(categorize_performance)
data2
```

```python
missing_values = data2.isnull().sum()
print(missing_values)
```

```python
rename_data = data2.rename(columns = {'Employee_ID':'ID'})
rename_data
```

```python
def tax_red(salary):
  return (salary*0.1)

data2['Tax'] = data2['Salary'].apply(tax_red)
data2
```
